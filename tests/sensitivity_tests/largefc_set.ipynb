```python
import os

input_file = "20210210_154121_S209-S-1-240min_Report.tsv"
samplemap_file = "samples.map.tsv"
results_dir = "results_w_ml"
results_dir_housekeeping_protnormed = "results_houskeeping_protnormed"
results_dir_no_norm = "results_w_ml_nonorm"
protein_subset_for_normalization_file="human_proteins.tsv"

spectronaut_file = "MP-LFC-OT.xlsx"
alphaquant_file = "./results_houskeeping_protnormed/S1_VS_S2.results.tsv"
mapping_file = "20210210_154121_S209-S-1-240min_Report.tsv"

run_diffanalysis = False

run_diffanalysis_nonorm = False

run_diffanalysis_housekeeping_protnormed = True

print(os.path.abspath("."))

housekeeping_prots = "housekeeping_proteins.tsv"
```

```python
if run_diffanalysis:
    import alphaquant.diff_analysis_manager as diffmgr

    diffmgr.run_pipeline(input_file=input_file, samplemap_file=samplemap_file, results_dir=results_dir, runtime_plots=True)
```

```python
if run_diffanalysis_housekeeping_protnormed:
    import alphaquant.diff_analysis_manager as diffmgr

    diffmgr.run_pipeline(input_file=input_file, samplemap_file=samplemap_file, results_dir=results_dir_housekeeping_protnormed, runtime_plots=True, protein_subset_for_normalization_file=housekeeping_prots, write_out_results_tree=False)
```

```python
if run_diffanalysis_nonorm:
    import alphaquant.diff_analysis_manager as diffmgr
    print(os.path.abspath("."))
    diffmgr.run_pipeline(input_file=input_file, samplemap_file=samplemap_file, results_dir=results_dir_no_norm, runtime_plots=True, normalize=True, use_iontree_if_possible=False)
```

```python
import alphaquant.benchmarking as aqbench
import alphaquant.visualizations as aqplot



fdr_thresold = 0.05

results_sn = aqbench.ResultsTableSpectronaut(spectronaut_file,input_name="Spectronaut",fdr_threshold=fdr_thresold)
results_aq = aqbench.ResultsTableAlphaQuant(alphaquant_file, input_name="AlphaQuant", fdr_threshold=fdr_thresold)


speciesannotator = aqbench.SpeciesAnnotator(mapping_file=mapping_file)
```

```python
merged_results = aqbench.MergedResultsTable([results_sn, results_aq])

speciesannotator.annotate_table_with_species(merged_results)

classplotter = aqbench.ClassificationBenchmarker(merged_results)
display(classplotter.speciescount_table)
ax = classplotter.plot_detected_proteins_per_species()
```
