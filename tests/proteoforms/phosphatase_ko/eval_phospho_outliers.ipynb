```python
import os

os.chdir("../../test_data/system_tests/proteoforms/mann_labs_phosphatases/RPAP2")
input_file = "proteome.RPAP2_SUBSET.tsv"
samplemap_file = "samples.map.RPAP2.proteome.tsv"
results_dir = "results_proteome"


input_file_phospho = "phospho.RPAP2_SUBSET.tsv.spectronaut_ptm_fragion_isotopes.aq_reformat.tsv"
samplemap_file_phospho = "samples.map.RPAP2.phospho.tsv"
results_dir_phospho = "results_phospho"
ml_info_table_phospho = "phospho.RPAP2_SUBSET.ml_info_table.tsv"


condpairs = [("RPAP2", "scramble")]
condpair = ("RPAP2", "scramble")

run_diffanalysis = False
run_diffanalysis_phospho = True

```

```python
import alphaquant.outlier_scoring as aqoutlier
import alphaquant.diffquant_utils as aqutils

condpairtree_proteome = aqutils.read_condpair_tree(cond1=condpair[0], cond2=condpair[1], results_folder=results_dir)
condpairtree_ptm = aqutils.read_condpair_tree(cond1=condpair[0], cond2=condpair[1], results_folder=results_dir_phospho)


```

```python
import alphaquant.outlier_scoring as aqoutlier


ptm_loader = aqoutlier.ModifiedPeptideLoader(condpairtree_ptm, specific_modification="_[")


outlier_loader = aqoutlier.OutlierPeptideLoader(condpairtree_proteome)

complemented_clusterloader = aqoutlier.ComplementedClusterLoader(outlier_peptide_loader=outlier_loader, modified_peptide_loader=ptm_loader)
```

```python
import matplotlib.pyplot as plt

quality_quantile = 0.2

filterconfigs = aqoutlier.ComplementedClusterFilterConfigs(ptm_abs_normfc_quantile=1, outlier_abs_normfc_quantile=0.5, outlier_quality_quantile=quality_quantile,modpep_quality_quantile=quality_quantile, ptm_absfc_quantile=1,number_mainclustpeps_quantile=0.7)


c_evaluator = aqoutlier.ComplementedClusterEvaluator(complemented_clusterloader,filterconfigs)
corr = c_evaluator.calculate_correlation()
print(f"correlation is {corr}")
ax = plt.axes()
c_evaluator.compare_regulation_directions(ax)

plt.show()

ax = plt.axes()
c_evaluator.scatter_fold_changes(ax)



#ax.set_xlim(-3, 3)
#ax.set_ylim(-3, 3)
plt.show()
```
