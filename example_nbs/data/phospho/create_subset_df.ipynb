```python
PHOSPHO_FILE = "/Users/constantin/workspace/alphaquant/test_data/system_tests/proteoforms/mann_labs_egfr/20230803_163821_PhosphoLiDIA_Report.tsv.zip"
PROTEOME_FILE = "/Users/constantin/workspace/alphaquant/test_data/system_tests/proteoforms/mann_labs_egfr/20230411_100022_20230330_Constantin_aQ2_Report.tsv"
```

```python
import pandas as pd
df_phospho = pd.read_csv(PHOSPHO_FILE, sep="\t")
df_proteome = pd.read_csv(PROTEOME_FILE, sep="\t")
```

```python
proteins_to_use = set(df_phospho["PG.ProteinGroups"].unique()).intersection(set(df_proteome["PG.ProteinGroups"].unique()))
print(proteins_to_use)
print(len(proteins_to_use))
```

```python
import numpy as np
sampled_proteins = list(np.random.choice(list(proteins_to_use), 30)) + ["P00533"]
```

```python
df_phospho_subet = df_phospho.set_index("PG.ProteinGroups").loc[sampled_proteins].reset_index()
df_proteome_subet = df_proteome.set_index("PG.ProteinGroups").loc[sampled_proteins].reset_index()

display(df_phospho_subet)
display(df_proteome_subet)
```

```python
FILE_FOLDER = "/Users/constantin/workspace/alphaquant/example_nbs/data/phospho"
```

```python
df_proteome_subet = df_proteome_subet[["_100ng" in filename for filename in df_proteome_subet["R.FileName"]]]
display(df_proteome_subet)
```

```python
df_phospho_subet.to_csv(FILE_FOLDER + "/phospho_subset.tsv", sep="\t", index=False)
df_proteome_subet.to_csv(FILE_FOLDER + "/proteome_subset.tsv", sep="\t", index=False)
```
